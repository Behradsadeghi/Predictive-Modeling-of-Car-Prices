```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv("/Users/behrad/Documents/imt tmrin/cardata.csv")
df= pd.DataFrame(data)
df
```

```python
print (pd.unique (df ["Fuel_Type"])) 
print (pd.unique (df["Seller_Type"])) 
print (pd.unique (df ["Transmission" ]))
print (pd.unique (df ["Owner"]))
```

```python
df.drop(['Car_Name'], axis=1, inplace=True)
df
```

```python
df['Age']= 2018 - df['Year']
df
```

```python
df.drop('Year',axis=1,inplace=True)
df
```

```python

df['Seller_Type'] = df['Seller_Type'].replace({'Dealer': 11, 'Individual': 12})


df['Transmission'] = df['Transmission'].replace({'Manual': 11, 'Automatic': 12})


df['Fuel_Type'] = df['Fuel_Type'].replace({'Petrol': 11, 'Diesel': 12, 'CNG': 13})

df
```

```python
x=df.drop("Selling_Price", axis=1)
y=df["Selling_Price"].values.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
regr=LinearRegression()
model=regr.fit(x_train,y_train)
```

```python
y_pred= regr.predict(x_test)
```

```python
print("Mean Absolute Error :",  metrics.mean_absolute_error(y_test,y_pred))
print("Mean Squared Error :", metrics.mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error :",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print("R2 Score:" , metrics.r2_score(y_test,y_pred))
result=regr.score(x_test,y_test)
result
```

```python

kf_validation=KFold(20)

result= cross_val_score(model,x,y,cv=kf_validation)
print(result)
```
